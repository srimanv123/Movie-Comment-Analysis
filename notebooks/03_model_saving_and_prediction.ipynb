```python
import pickle

with open("../models/final_logistic_model.pkl", "rb") as f:
    model = pickle.load(f)

with open("../models/tfidf_vectorizer.pkl", "rb") as f:
    vectorizer = pickle.load(f)

print("Model Loaded Successfully!")
```

```python
comments = [
    "cinema super bagundi",
    "worst movie waste of time",
    "excellent acting and direction",
    "average movie",
    "boring movie",
    "blockbuster movie",
    "songs are amazing",
    "movie was okay",
    "terrible direction",
    "must watch movie"
]

for comment in comments:
    prediction = model.predict(vectorizer.transform([comment]))
    print(f"{comment} --> {prediction[0]}")
```
