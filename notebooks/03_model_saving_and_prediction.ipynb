```python
import pickle

with open("../models/final_logistic_model.pkl", "rb") as f:
    model = pickle.load(f)

with open("../models/tfidf_vectorizer.pkl", "rb") as f:
    vectorizer = pickle.load(f)

print("Model Loaded Successfully!")
```

```python
comment = ["Movie super bagundi"]

vector = vectorizer.transform(comment)

prediction = model.predict(vector)

print("Prediction:", prediction[0])
```

```python
import numpy as np

comment = ["Movie super bagundi"]

vector = vectorizer.transform(comment)

prediction = model.predict(vector)

probabilities = model.predict_proba(vector)

confidence = np.max(probabilities) * 100

print("Prediction:", prediction[0])
print(f"Confidence: {confidence:.2f}%")
```

```python
import numpy as np

def predict_movie_rating(comment):

    vector = vectorizer.transform([comment])

    prediction = model.predict(vector)[0]

    confidence = np.max(model.predict_proba(vector)) * 100

    return prediction, confidence
```

```python
rating, confidence = predict_movie_rating(
    "Movie super bagundi"
)

print("Rating:", rating)
print(f"Confidence: {confidence:.2f}%")
```

```python
comments = [
    "Movie super bagundi",
    "Worst movie waste of time",
    "Excellent acting and direction",
    "Average movie",
    "Blockbuster movie",
    "Terrible screenplay",
    "Songs are amazing",
    "Movie is boring"
]

for comment in comments:
    rating, confidence = predict_movie_rating(comment)

    print(f"Comment    : {comment}")
    print(f"Prediction : {rating}")
    print(f"Confidence : {confidence:.2f}%")
    print("-" * 50)
```

```python
comments = [

    # Positive
    "Movie super bagundi",
    "Excellent acting and direction",
    "Blockbuster movie",
    "Must watch movie",
    "Loved every scene",

    # Negative
    "Worst movie waste of time",
    "Terrible screenplay",
    "Movie is boring",
    "Disappointing movie",
    "Very bad acting",

    # Neutral
    "Average movie",
    "Movie was okay",
    "One time watch",
    "Nothing special",

    # Romanized Telugu
    "Cinema bagundi",
    "Cinema waste",
    "Hero acting super",
    "Story antha bagaledu",

    # Mixed
    "First half was good but second half was boring",
    "Songs were excellent but story was weak"
]

for comment in comments:
    prediction = model.predict(vectorizer.transform([comment]))[0]

    print("="*60)
    print("Comment   :", comment)
    print("Prediction:", prediction)
```

```python
import numpy as np

for comment in comments:

    vector = vectorizer.transform([comment])

    prediction = model.predict(vector)[0]

    confidence = np.max(
        model.predict_proba(vector)
    ) * 100

    print("="*60)
    print("Comment    :", comment)
    print("Prediction :", prediction)
    print(f"Confidence : {confidence:.2f}%")
```

```python
import pandas as pd

comment = "Movie super bagundi"

vector = vectorizer.transform([comment])

prediction = model.predict(vector)[0]

probabilities = model.predict_proba(vector)[0]

classes = model.classes_

result = pd.DataFrame({
    "Class": classes,
    "Probability (%)": probabilities * 100
})

print("Prediction:", prediction)
print(result.sort_values("Probability (%)", ascending=False))
```
